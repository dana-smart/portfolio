### 3.3 Checking for Duplicate Values

```python
# Checking for duplicate values and counting them
orders_df.duplicated().sum()
```

```python
# Percentage of duplicate values
orders_df.duplicated().sum()/orders_df.shape[0]*100.0
```

From the above, there seems to be 178 duplicated rows in our dataset. This equates to ~1.5% of our data. While it seems like a small number, let's see if there's anything we can do other than deciding to remove these rows.

```python
# Let's take a closer look at the duplicate values
orders_df[orders_df.duplicated(keep=False)].head(40)
```

The duplicate rows appear to be within the same `Order_ID` for a particular order. This may indicate that customers are ordering an item more than once as a line item. My opinion is that some customers may have added the quantity they needed with the `Quantity` parameter while others have added the `Item_Name` as many times as they needed into the order. For this reason, I was originally going to keep the duplicate rows. 

After taking a look at the duplicate rows, it does in fact appear to be duplicate values. For example, order id 25496 should have a total of 4 products in the order: 2 plain papadums, 1 butter chicken and 1 pilau rice. However, we are seeing another record of the 2 plain papadums. If we include this second record, the number of total products wouldn't add up. In this case, the `Total_products` column helps us determine that these are duplicate rows. 

For this reason, I think we should remove the duplicate rows. This will be done in the next step. 

```python
# Remove duplicate values
orders_df.drop_duplicates(keep='first',inplace=True)
```

```python
# Checking for duplicates and counting
orders_df.duplicated().sum()
```

```python
# Check shape again
orders_df.shape
```

We no longer have duplicate values and can our new datafrrame consists of 119,005 rows and 6 columns. We can now check for missing data.

### 4.2 Order Date Column

```python
# Describe the Order_Date column
orders_df['Order_Date'].describe(datetime_is_numeric=True)
```

```python
# Count unique order dates
orders_df['Order_Date'].nunique()
print(f"Our dataframe has {orders_df['Order_Date'].nunique()} order dates.")
```

```python
# Find the min and max order date included in our dataset
print(orders_df['Order_Date'].min(), orders_df['Order_Date'].max())
```

The above tells us that the first order included in our dataset was placed on 2015-12-16 at 7:14pm and the last order was places on 2019-12-07 at 10:22pm. Were working with data within a four year period. 

From the above, we can note that the `Order_Date` consists of the date and time the order was placed. I think it would be helpful to split the date and time in order to work with the timestamps. For that reason, we will split these two elements and add them back to the dataframe in the next steps. I also want to extract the year, month and week day.

I was getting a timestamp error when splitting the order date, I was able to troubleshoot it on stack overflow with the use of df.apply(pd.Timestamp). Without it, I was not able convert the `Time` column format to datetime due to the missing time stamps represented as [00:00:00]. 

```python
# Extract year
orders_df['Year'] = orders_df['Order_Date'].dt.year

# Extract month
orders_df['Month'] = orders_df['Order_Date'].dt.month

# Extract day
orders_df['Day_of_Month'] = orders_df['Order_Date'].dt.day

# Extract hour
orders_df['Hour'] = orders_df['Order_Date'].dt.hour

# Extract day of week
orders_df['Day_of_Week'] = orders_df['Order_Date'].dt.day_name()
```

```python
# Look at the new dataframe
orders_df.head()
```

Let's check the data type for the new features.

```python
# Check data types
orders_df.info()
```

We have now created five additional columns with the date and time information. 

```python
# Group number of orders by year
orders_df['Order_Date'].groupby(orders_df.Order_Date.dt.year).agg('count')
```

```python
# Plot graph
sns.set_style("whitegrid")
plt.figure()
sns.countplot(x="Year", data=orders_df, color="steelblue")
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Orders by Year')
plt.show()
```

This gives us a snapshot of the orders placed based on the year. We find that the number orders increased year-over-year between 2016 and 2018. The biggest increase in sales was between 2016 and 2017, the increase was higher than 150%. This jump may be due to the restaurant establishing itself in the city. It reached a peak in 2018 before is started it's decline in orders. There is a big decrease (about 65% decrease) in order numbers between 2018 and 2019.
